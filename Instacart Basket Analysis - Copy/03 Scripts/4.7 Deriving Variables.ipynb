## 4. Creating a Price Label Column using Functions

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df.head()
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## 7. Creating a Column for Busiest Period of Day

```python
#Creating a List using if-then to add as a column
result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['busiest_day'].value_counts()
```

```python
#Creating a List using if-then to add as a column
results = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    results.append("Busiest days")
  elif value == 1:
    results.append("Busiest days")
  elif value == 3:
    results.append("Least busy days")
  elif value == 4:
    results.append("Least busy days")
  else:
    results.append("Regularly busy")
```

```python
ords_prods_merge['order_dow'].value_counts()
```

```python
ords_prods_merge['busiest_days'] = results
```
