## 4. Creating a max column for the max amount of order numbers grouped by user id/user

```python
ords_prods_merge.groupby('user_id').agg({'order_number' : ['max']})
```

```python
#Creating a max column for the max amount of order numbers grouped by user id/user 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking data
ords_prods_merge.head()
```

## 5. Creating loyalty flags using loc

```python
#Creating flags using loc
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking for the loaylty customer column
ords_prods_merge.head()
```

## 6. Checking basic statistics by loyalty status

```python
#checking basic statistics by loyalty status 
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean','min', 'max']})
```

## 7. Creating a column for the average price a user spends and spending flag

```python
#Creating a column for the average price a user spends
ords_prods_merge['average_price_of_customer_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.average)
```

```python
#checking data
ords_prods_merge.head()
```

```python
#creating a flag column for customer spending type 
ords_prods_merge.loc[ords_prods_merge['average_price_of_customer_spending'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['average_price_of_customer_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
pd.options.display.max_rows = None
```

```python
#checking data
ords_prods_merge.head(100)
```

```python
#doing a value coung
ords_prods_merge['spending_flag'].value_counts()
```

## 8. Creating a column for median of 'days since prior order' and frequency flag, and saving the prepared data as a pickle file. Also, creating a histogram for prices.

```python
#creating a column for median of 'days since prior order'
ords_prods_merge['prior_order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(10)
```

```python
#creating a flag column for frequency 
ords_prods_merge.loc[ords_prods_merge['prior_order_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['prior_order_median'] <= 20) & (ords_prods_merge['prior_order_median'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['prior_order_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge['frequency_flag'].value_counts()
```
