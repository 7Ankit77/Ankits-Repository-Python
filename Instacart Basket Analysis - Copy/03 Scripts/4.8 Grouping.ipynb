## 1. Importing necessary libraries: pandas and numpy

```python
#Importing Data
import pandas as pd
import numpy as np
import os
```

## 2. Importing data using pd.read_pickle() function

```python
#Importing dataframe
ords_prods_merge = pd.read_pickle(r'C:\Users\Ankit\Documents\CareerFoundry\Achievement 4\Instacart Basket Analysis\02 Data\Prepared Data\orders_products_merged.pkl')
```

## 3. Creating a data subset of the imported data

```python
#Creating Data Subset
df = ords_prods_merge[:1000000]
```

## 4. Checking the size and first 10 rows of the subset using .shape and .head() functions

```python
#checking the data out
df.shape
```

```python
df.head(10)
```

## 5. Grouping data by product name and department ID using .groupby() function

```python
#goruping using product name
df.groupby('product_name')
```

```python
#grouping data and showing average order number by department
df.groupby('department_id').agg({'order_number' : ['mean']})
```

## 6. Getting the average order number by department using .agg() and .mean() functions

```python
#getting mean from group as above using mean function
df.groupby('department_id')['order_number'].mean()
```

```python
#Gettting more descriptive statistics
df.groupby('department_id').agg({'order_number' : ['mean','min', 'max']})
```

## 7. Creating a new column to identify the maximum order number for each user using .groupby() and .transform() functions

```python
#creating a new column for max values
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## 8. Creating a column to flag customers' loyalty status based on their maximum order number using .loc() function and checking the counts of each type of customers classified by loyalty status using .value_counts() function. Finally, checking the result with a subset of the data using .head() function.

```python
#Creating a column to flag loyalty status for customers
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking counts of each type of customers classified by loyalty status
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
