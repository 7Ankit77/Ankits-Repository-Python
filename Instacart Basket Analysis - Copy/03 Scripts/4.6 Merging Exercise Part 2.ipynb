## 4. Dropping unnecessary columns from one of the datasets

```python
#Dropping unused columns for products_checked
products_checked = products_checked.drop(columns = ['Unnamed: 0'])
```

```python
products_checked.head()
```

```python
orders_products_combined.head()
```

```python
#dropping the merge column
orders_products_combined = orders_products_combined.drop(columns =['_merge', 'Unnamed: 0', 'Unnamed: 0.1'])
```

```python
#checking output after dropping merge column
orders_products_combined.head()
```

## 5. Merging the two datasets on a common column

```python
df_merged = products_checked.merge(orders_products_combined, on = 'product_id', indicator = True)
```

```python
df_merged.head()
```

```python
#Checking for consistency in the merged data
df_merged['_merge'].value_counts()
```
